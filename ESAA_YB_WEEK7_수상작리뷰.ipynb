**주제** 성격 유형(내향성/외향성) 예측

**데이터**

train: 한 사람에게서 발생한 다양한 정보를 기록한 웹로그
- id: 인덱스
- Time_spent_Alone: 혼자서 시간을 보내는 시간
- Stage_fear: 무대 공포증 유무
- Social_event_attendance: 사회 행사 참석 횟수
- Going_outside: 밖으로 나가는 횟수
- Drained_after_socializing: 사교 활동 후 소모되는지 여부
- Friends_circle_size: 친구 모임 규모
- Post_frequency
- Personality: Extrovert(외향)/Introvert(내향) <- TARGET

test: 한 사람에게서 발생한 다양한 정보를 기록한 웹로그

**코드 흐름**

(1) **EDA**

Yes/No라고 답변하는 경우를 1/0으로 map
- Stage_fear, Drained_after_socializing, Personality

전체 칼럼을 대상으로 하는 상관 히트맵을 만듦

서로 다른 변수 사이의 비선형 관계 탐지
- mutual_info_regression

(2) **데이터 전처리**

중복된 행 제거 - 같은 사람이 중복 기입하거나 데이터가 여러 번 들어간 경우를 제거하는 작업

데이터 병합 - 왜 하는지 잘 모르겠다..

범주형 변수 처리
- 결측값을 "missing"으로 채워 넣음
- 레이블 인코딩 - Extrovert와 Introvert를 0, 1로 map

(3) **기본 모델 학습**

CatBoost, XGBoost, HistGradientBoostingClassifier, LightGBM(gbdt, goss, dart), AutoGluon을 이용

(4) **Ensembling**

LogisticRegression, Weighted average를 이용

**차별점 및 배울점**

상관 히트맵분 아니라 mutual_info_regression을 이용해 비선형 관계를 탐지했다. 선형 관계만으로는 알기 힘든 요소들을 이를 통해 알 수 있게 해 모델 학습에 도움을 주었다.

데이터가 여러 번 들어간 경우에 대해 생각해 본 적이 없었는데, 이 수상작을 리뷰하며 중복된 행이 있을 수도 있다는 생각이 들었다. 데이터 전처리 과정 중에 참고해야겠다.

결측값을 "missing"이라고 입력해 새로운 범주로 인식하게 했는데, 이에 대해 왜 이렇게 했는지 이해가 잘 되지 않는다. 결측값에 있어 "missing"에도 외향성과 내향성을 구분할 수 있다고 판단한 건지, 이 데이터에 있어 이 점이 중요한 요소로 여겨지는지 잘 모르겠다.
