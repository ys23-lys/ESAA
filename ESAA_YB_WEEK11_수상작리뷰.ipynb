**주제** 금 가격 예측

**데이터**
- Date: 날짜
- Open: 오픈프라이스
- High: 가장 높은 금액
- Low: 가장 낮은 금액
- Close: 마지막 거래 가격(종가)
- Volume: 용량

**코드 흐름**

(1) **전처리**

날짜 형식 변경 <- 인덱스로 변경

시간 열 삭제

결측값의 경우 이전 날짜의 데이터로 채우기

새로운 변수 만들기(var_Prev_Open, var_Prev_Close, var_Prev_High, var_Prev_Low, var_Close, prev_Close 등..)
- var_Prev_: df[''].shift(1)-df[''].shift(2): 전날 값 - 전전날 값(전일의 변동성을 알려주는 변수)
- var_Close: df['Close'] - df['Close'].shift(1): 오늘 종가 - 전날 종가(오늘 가격이 얼마나 올랐는지/떨어졌는지를 알려주는 변수) **Target**
- prev_Close: 전날 종가를 저장하는 함수
- 1~30일의 과거의 변화량을 생성해 feature로 사용

(2) **시각화**

train 자료와 test 자료를 넣어 시계열 시각화를 진행

(3) **모델 학습**

ForecasterAutoreg 이용, 회귀 모델은 DecisionTreeRegressor
- 지난 30일의 데이터를 입력해 그 다음 날의 데이터를 예측하도록 학습

GridSearch로 최적 하이퍼 파라미터를 찾음

모델, 목표변수(var_Close), 외생변수(var_Prev_변수), 하이퍼 파라미터 전달

**시각화** 관측값과 예측값을 겹쳐 시계열 시각화를 진행

**차별점 및 배울점**

Target값이 따로 주어져 있지 않아 직접 이 변수를 만들어 이를 적용해야 하는 경우도 있다는 것을 알았다. 이때 무엇을 예측할지 아는 것이 중요한데, (오늘 종가 - 전날 종가)를 변수로 만들어 오늘의 가격 상승/하락 예측을 진행했다는 점에서 전처리 전 피처들을 보다 주의 깊게 살펴야겠다는 생각이 들었다.

Target값뿐만 아니라 var_Prev_, Prev_Close_ 등 필요한 변수를 직접 만들어 이용했는데, 지금까지는 Target값과 피처들이 정제되어 주어진 데이터에서 모델 학습을 진행했다면, 보다 원시적인 데이터를 가공하고 처리하는 예시를 볼 수 있었다고 생각한다.
