**차별점 및 배울점**

결측치가 0일 경우를 고려해 데이터 처리 과정에서 data1과 data2로 그룹을 분할하여 수행하였다는 점에서 인상깊었다.

pairplot을 만들어 전체 데이터의 시각화를 한눈에 더욱 쉽게 알아볼 수 있었다. 특히 이 데이터의 경우 피처의 수가 많아 일일이 피처를 비교하기 힘들기 때문에 한번에 정리하여 타깃값에 영향을 미치는 피처를 알아내기에 더욱 적합했다고 생각한다.

피처를 시각화하고 가공하는 데서 그치지 않고 계속해서 타깃값에 영향을 미칠 것 같은 피처를 새롭게 만들어 유의미한 피처로 이용하는 점이 인상깊었다.

유의미한 변수를 선택하기 위해 RFECV 함수를 이용했는데, 피처 중요도를 시각화하여 파악하는 데서 그치지 않고 수많은 피처들이 계속 추가됨에도 불구하고 이 함수를 이용해 간편하게 유의미한 피처를 분류할 수 있었다.
