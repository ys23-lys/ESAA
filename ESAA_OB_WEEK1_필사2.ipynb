```python
from sklearn.datasets import fetch_20newsgroups

# subset='train'мңјлЎң н•ҷмҠөмҡ© лҚ°мқҙн„°л§Ң м¶”м¶ң, remove=('headers','footers','quotes')лЎң лӮҙмҡ©л§Ң м¶”м¶ң
train_news=fetch_20newsgroups(subset='train',remove=('headers','footers','quotes'),random_state=156)
X_train=train_news.data
y_train=train_news.target

# subset='test'мңјлЎң н…ҢмҠӨнҠё лҚ°мқҙн„°л§Ң м¶”м¶ң, remove=('headers','footers','quotes')лЎң лӮҙмҡ©л§Ң м¶”м¶ң
test_news=fetch_20newsgroups(subset='test',remove=('headers','footers','quotes'),random_state=156)
X_test=test_news.data
y_test=test_news.target
print('н•ҷмҠө лҚ°мқҙн„° нҒ¬кё° {0}, н…ҢмҠӨнҠё лҚ°мқҙн„° нҒ¬кё° {1}'.format(len(train_news.data),
                                             len(test_news.data)))
```

### н”јмІҳ лІЎн„°нҷ” ліҖнҷҳкіј лЁёмӢ лҹ¬лӢқ лӘЁлҚё н•ҷмҠө/мҳҲмёЎ/нҸүк°Җ

н…ҢмҠӨнҠё лҚ°мқҙн„°м—җм„ң CountVectorizerлҘј м Ғмҡ©н•  л•ҢлҠ” **н•ҷмҠө лҚ°мқҙн„°лҘј мқҙмҡ©н•ҙ fit()мқҙ мҲҳн–үлҗң CountVectorizer к°қмІҙлҘј мқҙмҡ©н•ҙ** н…ҢмҠӨнҠё лҚ°мқҙн„°лҘј transfromн•ҙм•ј н•Ё.
- н•ҷмҠө мӢң м„Өм •лҗң CountVectorizerмқҳ н”јмІҳ к°ңмҲҳмҷҖ н…ҢмҠӨнҠё лҚ°мқҙн„°лҘј CountVectorizerлЎң ліҖнҷҳн•  н”јмІҳ к°ңмҲҳк°Җ к°ҷм•„м§җ.

н…ҢмҠӨнҠё лҚ°мқҙн„°лҘј н”јмІҳ лІЎн„°нҷ”н•  л•Ң **fit_transfrom()мқ„ мӮ¬мҡ©н•ҳл©ҙ м•Ҳ лҗЁ**

н…ҚмҠӨнҠё л¶„м„қм—җм„ң лЁёмӢ лҹ¬лӢқ лӘЁлҚёмқҳ м„ұлҠҘмқ„ н–ҘмғҒмӢңнӮӨлҠ” л°©лІ•
1. мөңм Ғмқҳ ML м•Ңкі лҰ¬мҰҳ м„ нғқ
2. мөңмғҒмқҳ н”јмІҳ м „мІҳлҰ¬ мҲҳн–ү

```python
from sklearn.feature_extraction.text import CountVectorizer

# Count VectorizationмңјлЎң н”јмІҳ лІЎн„°нҷ” ліҖнҷҳ мҲҳн–ү.
cnt_vect=CountVectorizer()
cnt_vect.fit(X_train)
X_train_cnt_vect=cnt_vect.transform(X_train)

# н•ҷмҠө лҚ°мқҙн„°лЎң fit()лҗң CountVectorizerлҘј мқҙмҡ©н•ҙ н…ҢмҠӨнҠё лҚ°мқҙн„°лҘј н”јмІҳ лІЎн„°нҷ” ліҖнҷҳ мҲҳн–ү.
X_test_cnt_vect=cnt_vect.transform(X_test)

print('н•ҷмҠө лҚ°мқҙн„° н…ҚмҠӨнҠёмқҳ CountVectorizer Shape:',X_train_cnt_vect.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')

# LogisticRegressionмқ„ мқҙмҡ©н•ҳм—¬ н•ҷмҠө/мҳҲмёЎ/нҸүк°Җ мҲҳн–ү.
lr_clf=LogisticRegression(solver='liblinear')
lr_clf.fit(X_train_cnt_vect,y_train)
pred=lr_clf.predict(X_test_cnt_vect)
print('CountVectorized Logistic Regressionмқҳ м—җмёЎ м •нҷ•лҸ„лҠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF лІЎн„°нҷ”лҘј м Ғмҡ©н•ҙ н•ҷмҠө лҚ°мқҙн„° м„ёнҠёмҷҖ н…ҢмҠӨнҠё лҚ°мқҙн„° м„ёнҠё ліҖнҷҳ.
tfidf_vect=TfidfVectorizer()
tfidf_vect.fit(X_train)
X_train_tfidf_vect=tfidf_vect.transform(X_train)
X_test_tfidf_vect=tfidf_vect.transform(X_test)

# LogisticRegressionмқ„ мқҙмҡ©н•ҙ н•ҷмҠө/мҳҲмёЎ/нҸүк°Җ мҲҳн–ү.
lr_clf=LogisticRegression(solver='liblinear')
lr_clf.fit(X_train_tfidf_vect,y_train)
pred=lr_clf.predict(X_test_tfidf_vect)
print('TF-IDF Logistic Regressionмқҳ мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

```python
# stop words н•„н„°л§Ғмқ„ м¶”к°Җн•ҳкі  ngramмқ„ кё°ліё (1,1)м—җм„ң (1,2)лЎң ліҖкІҪн•ҙ н”јмІҳ лІЎн„°нҷ” м Ғмҡ©.
tfidf_vect=TfidfVectorizer(stop_words='english',ngram_range=(1,2),max_df=300)
tfidf_vect.fit(X_train)
X_train_tfidf_vect=tfidf_vect.transform(X_train)
X_test_tfidf_vect=tfidf_vect.transform(X_test)

lr_clf=LogisticRegression(solver='liblinear')
lr_clf.fit(X_train_tfidf_vect,y_train)
pred=lr_clf.predict(X_test_tfidf_vect)
print('TF-IDF Vectorized Logistic Regressionмқҳ мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

```python
from sklearn.model_selection import GridSearchCV

# мөңм Ғ Cк°’ лҸ„м¶ң нҠңлӢқ мҲҳн–ү. CVлҠ” 3 нҸҙл“ң м„ёнҠёлЎң м„Өм •.
params={'C':[0.01,0.1,1,5,10]}
grid_cv_lr=GridSearchCV(lr_clf,param_grid=params,cv=3,scoring='accuracy',verbose=1)
grid_cv_lr.fit(X_train_tfidf_vect,y_train)
print('Logistic Regression best C parameter:',grid_cv_lr.best_params_)

# мөңм Ғ C к°’мңјлЎң н•ҷмҠөлҗң grid_cvлЎң мҳҲмёЎ л°Ҹ м •нҷ•лҸ„ нҸүк°Җ.
pred=grid_cv_lr.predict(X_test_tfidf_vect)
print('TF-IDF Vectorized Logistic Regressionмқҳ мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

### мӮ¬мқҙнӮ·лҹ° нҢҢмқҙн”„лқјмқё(Pipeline) мӮ¬мҡ© л°Ҹ GridSearchCVмҷҖмқҳ кІ°н•©

Pipeline: лҚ°мқҙн„°мқҳ к°Җкіө, ліҖнҷҳ л“ұмқҳ м „мІҳлҰ¬мҷҖ м•Ңкі лҰ¬мҰҳ м Ғмҡ©мқ„ н•ңкәјлІҲм—җ мҠӨнҠёлҰј кё°л°ҳмңјлЎң мІҳлҰ¬. - м§ҒкҙҖм Ғмқё ML лӘЁлҚё мҪ”л“ң мғқм„ұ. мҲҳн–ү мӢңк°„ м Ҳм•Ҫ.

```python
from sklearn.pipeline import Pipeline
pipeline=Pipeline([('tfidf_vect',TfidfVectorizer(stop_words='english')),
                   ('lr_clf',LogisticRegression(random_state=156))])
```

```python
from sklearn.pipeline import Pipeline

# TfidVectorizer к°қмІҙлҘј tfidf_vectлЎң, LogisticRegression к°қмІҙлҘј lr_clfлЎң мғқм„ұн•ҳлҠ” Pipeline мғқм„ұ
pipeline=Pipeline([
    ('tfidf_vect',TfidfVectorizer(stop_words='english',ngram_range=(1,2),max_df=300)),
    ('lr_clf',LogisticRegression(solver='liblinear',C=10))
])

# лі„лҸ„мқҳ TfidVectorizer к°қмІҙмқҳ fit(), transform()кіј LogisticRegressionмқҳ fit(), predict()к°Җ н•„мҡ” м—ҶмқҢ.
# pipelineмқҳ fit()кіј predict()л§ҢмңјлЎң н•ңкәјлІҲм—җ н”јмІҳ лІЎн„°нҷ”мҷҖ ML н•ҷмҠө/м—җмёЎмқҙ к°ҖлҠҘ.
pipeline.fit(X_train,y_train)
pred=pipeline.predict(X_test)
print('Pipelineмқ„ нҶөн•ң Logistic Regressionмқҳ мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

pipeline=Pipeline([
    ('tfidf_vect',TfidfVectorizer(stop_words='english')),
    ('lr_clf',LogisticRegression(solver='liblinear'))
])

# Pipelineм—җ кё°мҲ лҗң к°Ғк°Ғмқҳ к°қмІҙ ліҖмҲҳм—җ м–ёлҚ”л°”(_) 2к°ңлҘј м—°лӢ¬м•„ л¶ҷм—¬ GridSearchCVм—җ мӮ¬мҡ©лҗ  нҢҢлқјлҜён„°/н•ҳмқҙнҚјнҢҢлқјлҜён„° мқҙлҰ„кіј к°’мқ„ м„Өм •
params={'tfidf_vect__ngram_range':[(1,1),(1,2),(1,3)],
        'tfidf_vect__max_df':[100,300,700],
        'lr_clf__C':[1,5,10]}

# GridSearchCVмқҳ мғқм„ұмһҗм—җ Estimatorк°Җ м•„лӢҢ Pipeline к°қмІҙ мһ…л Ҙ
grid_cv_pipe=GridSearchCV(pipeline,param_grid=params,cv=3,scoring='accuracy',verbose=1)
grid_cv_pipe.fit(X_train,y_train)
print(grid_cv_pipe.best_params_,grid_cv_pipe.best_score_)

pred=grid_cv_pipe.predict(X_test)
print('Pipelineмқ„ нҶөн•ң Logistic Regressionмқҳ мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.3f}'.format(accuracy_score(y_test,pred)))
```

## 05 к°җм„ұ л¶„м„қ

### к°җм„ұ л¶„м„қ мҶҢк°ң

к°җм„ұ л¶„м„қ(Sentiment Analysis): л¬ём„ңмқҳ мЈјкҙҖм Ғмқё к°җм„ұ/мқҳкІ¬/к°җм •/кё°л¶„ л“ұмқ„ нҢҢм•…н•ҳкё° мң„н•ң л°©лІ•. л¬ём„ң лӮҙ н…ҚмҠӨнҠёк°Җ лӮҳнғҖлӮҙлҠ” мЈјкҙҖм Ғ лӢЁм–ҙмҷҖ л¬ёл§Ҙмқ„ кё°л°ҳмңјлЎң к°җм„ұ(Sentiment) мҲҳм№ҳлҘј кі„мӮ°н•ҳлҠ” л°©лІ•. - кёҚм • к°җм„ұ м§ҖмҲҳ, л¶Җм • к°җм„ұ м§ҖмҲҳлҘј н•©мӮ°н•ҙ кёҚм • к°җм„ұ, л¶Җм • к°җм„ұ кІ°м •.
- м§ҖлҸ„н•ҷмҠө: н•ҷмҠө лҚ°мқҙн„°мҷҖ нғҖк№ғ л Ҳмқҙлё” к°’мқ„ кё°л°ҳмңјлЎң н•ҷмҠө мҲҳн–ү. лӢӨлҘё лҚ°мқҙн„°мқҳ к°җм„ұ л¶„м„қмқ„ мҳҲмёЎ.
- л№„м§ҖлҸ„н•ҷмҠө: **'Lexicon'**мқҙлқјлҠ” к°җм„ұ м–ҙнңҳ мӮ¬м „ мқҙмҡ©.

### м§ҖлҸ„н•ҷмҠө кё°л°ҳ к°җм„ұ л¶„м„қ мӢӨмҠө - IMDB мҳҒнҷ”нҸү

```python
import pandas as pd

review_df=pd.read_csv("/content/drive/MyDrive/Colab/ESAA/notebook/labeledTrainData.tsv",sep="\t",quoting=3)
review_df.head(3)
```

```python
print(review_df['review'][0])
```

```python
import re

# <br> html нғңк·ёлҠ” replace н•ЁмҲҳлЎң кіөл°ұмңјлЎң ліҖнҷҳ
review_df['review']=review_df['review'].str.replace('<br />','')

# нҢҢмқҙмҚ¬мқҳ м •к·ң н‘ңнҳ„мӢқ лӘЁл“Ҳмқё reлҘј мқҙмҡ©н•ҙ мҳҒм–ҙ л¬ёмһҗм—ҙмқҙ м•„лӢҢ л¬ёмһҗлҠ” лӘЁл‘җ кіөл°ұмңјлЎң ліҖнҷҳ
review_df['reiview']=review_df['review'].apply(lambda x: re.sub("[^a-zA-Z]"," ",x))
```

```python
from sklearn.model_selection import train_test_split

class_df=review_df['sentiment']
feature_df=review_df.drop(['id','sentiment'],axis=1,inplace=False)
X_train,X_test,y_train,y_test=train_test_split(feature_df,class_df,test_size=0.3,random_state=156)
X_train.shape,X_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,roc_auc_score

# мҠӨнҶұ мӣҢл“ңлҠ” English, ngramмқҖ (1,2)лЎң м„Өм •н•ҙ CountVectorization мҲҳн–ү
# LogisticRegressionмқҳ CлҠ” 10мңјлЎң м„Өм •.
pipeline=Pipeline([
    ('cnt_vect',CountVectorizer(stop_words='english',ngram_range=(1,2))),
    ('lr_clf',LogisticRegression(solver='liblinear',C=10))
])

# Pipeline к°қмІҙлҘј мқҙмҡ©н•ҙ fit(), predict()лЎң н•ҷмҠө/мҳҲмёЎ мҲҳн–ү. predict_proba()лҠ” roc_auc л•Ңл¬ём—җ мҲҳн–ү.
pipeline.fit(X_train['review'],y_train)
pred=pipeline.predict(X_test['review'])
pred_probs=pipeline.predict_proba(X_test['review'])[:,1]

print('мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.4f}, ROC-AUCлҠ” {1:.4f}'.format(accuracy_score(y_test,pred),roc_auc_score(y_test,pred_probs)))
```

```python
# мҠӨнҶұ мӣҢл“ңлҠ” english, filtering, ngramмқҖ (1,2)лЎң м„Өм •н•ҙ TF-IDF лІЎн„°нҷ” мҲҳн–ү
# LogisticRegressionмқҳ CлҠ” 10мңјлЎң м„Өм •
pipeline=Pipeline([
    ('tfidf_vect',TfidfVectorizer(stop_words='english',ngram_range=(1,2))),
    ('lr_clf',LogisticRegression(solver='liblinear',C=10))
])

pipeline.fit(X_train['review'],y_train)
pred=pipeline.predict(X_test['review'])
pred_probs=pipeline.predict_proba(X_test['review'])[:,1]

print('мҳҲмёЎ м •нҷ•лҸ„лҠ” {0:.4f}'.format(accuracy_score(y_test,pred),
                               roc_auc_score(y_test,pred_probs)))
```

### л№„м§ҖлҸ„н•ҷмҠө кё°л°ҳ к°җм„ұ л¶„м„қ мҶҢк°ң

**Lexicon**: к°җм„ұл§Ңмқ„ л¶„м„қн•ҳкё° мң„н•ҙ м§Җмӣҗн•ҳлҠ” к°җм„ұ м–ҙнңҳ мӮ¬м „.

к°җм • м§ҖмҲҳ(Polarity score): кёҚм • к°җм„ұ, л¶Җм • к°җм„ұмқҳ м •лҸ„лҘј мқҳлҜён•ҳлҠ” мҲҳм№ҳ.

**NLTK** нҢЁнӮӨм§Җк°Җ к°җм„ұ мӮ¬м „мқ„ кө¬нҳ„.
- WordNet: мӢңл§ЁнӢұ(semantic: л¬ёл§ҘмғҒ мқҳлҜё) л¶„м„қмқ„ м ңкіөн•ҳлҠ” м–ҙнңҳ мӮ¬м „.
   - SentiWordNet: WordNetмқҳ Synset к°ңл…җмқ„ к°җм„ұ л¶„м„қм—җ м Ғмҡ©.
   - VADER: мҶҢм…ң лҜёл””м–ҙ н…ҚмҠӨнҠём—җ лҢҖн•ң к°җм„ұ л¶„м„қмқ„ м ңкіөн•ҳкё° мң„н•ң нҢЁнӮӨм§Җ. л№ лҘё мҲҳн–ү мӢңк°„мқ„ ліҙмһҘн•Ё.
   - Pattern: мҳҲмёЎ м„ұлҠҘмқ„ ліҙмһҘн•Ё.


#### SentiWordNetмқ„ мқҙмҡ©н•ң к°җм„ұ л¶„м„қ

present.n.01

- n: лӘ…мӮ¬ н’ҲмӮ¬
- 01: лӘ…мӮ¬лЎңм„ң к°Җм§ҖлҠ” м—¬лҹ¬ к°Җм§Җ мқҳлҜёлҘј кө¬л¶„н•ҳлҠ” мқёлҚұмҠӨ

```python
# synset к°қмІҙлҘј лӢЁм–ҙлі„лЎң мғқм„ұн•©лӢҲлӢӨ.
tree=wn.synset('tree.n.01')
lion=wn.synset('lion.n.01')
tiger=wn.synset('tiger.n.02')
cat=wn.synset('cat.n.01')
dog=wn.synset('dog.n.01')

entities=[tree,lion,tiger,cat,dog]
similarities=[]
entity_names=[entity.name().split('.')[0] for entity in entities]

# лӢЁм–ҙлі„ synsetмқ„ л°ҳліөн•ҳл©ҙм„ң лӢӨлҘё лӢЁм–ҙмқҳ synsetкіј мң мӮ¬лҸ„лҘј мёЎм •н•©лӢҲлӢӨ.
for entity in entities:
  similarity=[round(entity.path_similarity(compared_entity),2)
              for compared_entity in entities]
  similarities.append(similarity)

# к°ңлі„ лӢЁм–ҙлі„ synsetкіј лӢӨлҘё лӢЁм–ҙмқҳ synsetкіјмқҳ мң мӮ¬лҸ„лҘј DataFrame нҳ•нғңлЎң м ҖмһҘн•©лӢҲлӢӨ.
similarity_df=pd.DataFrame(similarities,columns=entity_names,index=entity_names)
similarity_df
```

##### SentiWordNetмқ„ мқҙмҡ©н•ң мҳҒнҷ” к°җмғҒнҸү к°җм„ұ л¶„м„қ

1. л¬ём„ңлҘј л¬ёмһҘ лӢЁмң„лЎң л¶„н•ҙ
2. л¬ёмһҘмқ„ лӢЁм–ҙ лӢЁмң„лЎң нҶ нҒ°нҷ”н•ҳкі  н’ҲмӮ¬ нғңк№…
3. н’ҲмӮ¬ нғңк№…лҗң лӢЁм–ҙ кё°л°ҳмңјлЎң synset к°қмІҙмҷҖ senti_synset к°қмІҙлҘј мғқм„ұ
4. senti_synset к°қмІҙм—җм„ң кёҚм • к°җм„ұ/л¶Җм • к°җм„ұ м§ҖмҲҳлҘј кө¬н•ҳкі  н•©мӮ°н•ҙ нҠ№м • мһ„кі„м№ҳ к°’ мқҙмғҒмқј л•Ң кёҚм • к°җм„ұмңјлЎң, к·ёл Үм§Җ м•Ҡмқ„ л•Ң л¶Җм • к°җм„ұмңјлЎң кІ°м •.

```python
from nltk.corpus import wordnet as wn

# к°„лӢЁн•ң NTLK PenTreebank TagлҘј кё°л°ҳмңјлЎң WordNet кё°л°ҳмқҳ н’ҲмӮ¬ TagлЎң ліҖнҷҳ
def penn_to_wn(tag):
  if tag.startswith('J'):
    return wn.ADJ
  elif tag.startswith('N'):
    return wn.NOUN
  elif tag.startswith('R'):
    return wn.ADV
  elif tag.startswith('V'):
    return wn.VERB
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import sentiwordnet as swn
from nltk import sent_tokenize,word_tokenize,pos_tag

def swn_polarity(text):
  # к°җм„ұ м§ҖмҲҳ мҙҲкё°нҷ”
  sentiment=0.0
  tokens_count=0

  lemmatizer=WordNetLemmatizer()
  raw_sentences=sent_tokenize(text)
  # л¶„н•ҙлҗң л¬ёмһҘлі„лЎң лӢЁм–ҙ нҶ нҒ° -> н’ҲмӮ¬ нғңк№… нӣ„м—җ SentiSynset мғқм„ұ -> к°җм„ұ м§ҖмҲҳ н•©мӮ°
  for raw_sentence in raw_sentences:
    # NTLK кё°л°ҳмқҳ н’ҲмӮ¬ нғңк№… л¬ёмһҘ м¶”м¶ң
    tagged_sentence=pos_tag(word_tokenize(raw_sentence))
    for word,tag in tagged_sentence:

      # WordNet кё°л°ҳ н’ҲмӮ¬ нғңк№…кіј м–ҙк·ј м¶”м¶ң
      wn_tag=penn_to_wn(tag)
      if wn_tag not in (wn.NOUN,wn.ADJ,wn.ADV):
        continue
      lemma=lemmatizer.lemmatize(word,pos=wn_tag)
      if not lemma:
        continue
      # м–ҙк·јмқ„ м¶”м¶ңн•ң лӢЁм–ҙмҷҖ WordNet кё°л°ҳ н’ҲмӮ¬ нғңк№…мқ„ мһ…л Ҙн•ҙ Synset к°қмІҙлҘј мғқм„ұ.
      synsets=wn.synsets(lemma,pos=wn_tag)
      if not synsets:
        continue
      # sentiwordnetмқҳ к°җм„ұ лӢЁм–ҙ л¶„м„қмңјлЎң к°җм„ұ synset м¶”м¶ң
      # лӘЁл“  лӢЁм–ҙм—җ лҢҖн•ҙ кёҚм • к°җм„ұ м§ҖмҲҳлҠ” +лЎң л¶Җм • к°җм„ұ м§ҖмҲҳлҠ” -лЎң н•©мӮ°н•ҙ к°җм„ұ м§ҖмҲҳ кі„мӮ°
      synset=synsets[0]
      swn_synset=swn.senti_synset(synset.name())
      sentiment+=(swn_synset.pos_score()-swn_synset.neg_score())
      tokens_count+=1

  if not tokens_count:
    return 0

  # мҙқ scoreк°Җ 0 мқҙмғҒмқј кІҪмҡ° кёҚм •(Positive) 1, к·ёл Үм§Җ м•Ҡмқ„ кІҪмҡ° л¶Җм •(Negative) 0 л°ҳнҷҳ
  if sentiment >= 0:
    return 1

  return 0
```

```python
def vader_polarity(review,threshold=0.1):
  analyzer=SentimentIntensityAnalyzer()
  scores=analyzer.polarity_scores(review)

  # compound к°’м—җ кё°л°ҳн•ҙ threshold мһ…л Ҙк°’ліҙлӢӨ нҒ¬л©ҙ 1, к·ёл Үм§Җ м•Ҡмңјл©ҙ 0мқ„ л°ҳнҷҳ
  agg_score=scores['compound']
  final_sentiment=1 if agg_score >=threshold else 0
  return final_sentiment

# apply lambda мӢқмқ„ мқҙмҡ©н•ҙ л ҲмҪ”л“ңлі„лЎң vader_polarity()лҘј мҲҳн–үн•ҳкі  кІ°кіјлҘј 'vader_preds'м—җ м ҖмһҘ
review_df['vader_preds']=review_df['review'].apply(lambda x: vader_polarity(x,0.1))
y_target=review_df['sentiment'].values
vader_preds=review_df['vader_preds'].values

print(confusion_matrix(y_target,vader_preds))
print("м •нҷ•лҸ„:",np.round(accuracy_score(y_target,vader_preds),4))
print("м •л°ҖлҸ„:",np.round(precision_score(y_target,vader_preds),4))
print("мһ¬нҳ„мңЁ:",np.round(recall_score(y_target,vader_preds),4))
```
