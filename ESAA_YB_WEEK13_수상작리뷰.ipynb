**주제** 칼로리 소모량 예측 AI 해커톤

**데이터**
- ID
- Exercise_duration: 운동 시간(분)
- Body_temperature(F)
- BPM: 심박수
- Height(Feet): 키(피트)
- Height(Remainder_Inches): 키(피트 계산 후 더해야 할 키)
- Weight(lb): 몸무게(파운드)
- Weight_Status: 체중 상태
- Gender
- Age
- Calories_Burned: 칼로리 소모량(목표 예측값)

**코드 흐름**

(1) **시각화**

Exercise_Duration, BPM, Age, Weight의 상관 히트맵 그리기

(2) **전처리**

Gender(범주형 변수)를 숫자로 인코딩 - 라벨 인코딩 이용.

(3) **모델링**

**LinearRegression**으로 회귀 모델 생성/예측/mse로 평가

**Ridge**로 회귀 모델 생성/예측/mse로 평가

(4) **스태킹**

3차 다항 특성 생성기(polynomialFeatures(dgree=3): 변수 간 상호작용 항만 생성
- Train 데이터를 3차 상호작용 특성으로 변환

+ (3)에서 만든 선형회귀 모델과 릿지 회귀 모델을 생성해 예측값 생성.

**차별점 및 배울점**

전처리 과정에서 타깃값에 영향을 미치는 피처들만 추출해 train 데이터에 넣었다.

상관 히트맵을 만들었을 때 BPM과 타깃값의 상관계수가 매우 높게 나왔는데, 이에 대한 처리가 보다 필요했으리라고 생각한다.
- 연관성이 너무 높기 때문에 BPM 피처를 삭제하거나 살펴보는 방식이 필요했을 것이다.

전처리 시 train 데이터와 test 데이터에 있는 값을 비교해 test에만 있는 값 때문에 에러가 나는 문제가 있는데, 이를 예방하는 코드를 작성했다.
