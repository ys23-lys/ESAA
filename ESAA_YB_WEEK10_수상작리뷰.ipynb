**주제** 건강 보험료 예측

**데이터**
- age: 나이
- sex: 성별
- bmi: 체질량 지수
- children: 아이 수
- smoker: 흡연 유무
- region: 지역(southeast/southwest/others)
- charges: 건강 보험료 <- TARGET

**코드 흐름**

(1) **EDA**

head, shape, info, describe 이용

데이터 시각화
- age, bmi, children, charges를 histogram으로 나타냄.
- sex, smoker, region을 bar 그래프로 나타냄.(**범주형**)
- scatter plot으로 피처 간 관계 나타냄.
   - age vs charges
   - bmi vs charges
   - children vs charges
- boxplot으로 smoker와 charges 관계 나타냄.(범주형)
- correlation heatmap으로 범주형을 제외한 피처 간의 관계 나타냄.

(2) **모델 학습**

target과 features 구분.

학습/검증/테스트 데이터를 60/20/20으로 분할.

**데이터 전처리**
- num: 숫자형 데이터 처리 - 표준화
- cat: 범주형 데이터 처리 - 원핫 인코딩

LinearRegression/RandomForestRegressor/GradientBoostingRegressor 이용해 모델 학습.

MAE RMSE R2 평가 지표로 사용.

**차별점 및 배울점**

데이터 전처리 -> EDA -> 모델 학습 순서로 진행한다고 생각했는데, EDA를 먼저 진행한 후 모델 학습 직전에 필요한 데이터만 전처리하면 과정이 보다 줄어든다는 점을 깨달았다.

데이터 시각화를 진행했을 때, 상관 히트맵으로 선형 상관관계에 대해서 알아보았으나 대부분 유의미한 상관계수가 나오지 않았다. 즉 이뿐으로는 건강 보험료의 예측이 힘들다는 점을 알 수 있는데, 선형 관계만을 나타낸 것이었으므로 비선형 관계를 함께 나타내어 봤으면 유의미한 피처를 식별하는 데 더욱 도움이 되었을 것 같다.

데이터 전처리를 진행할 때 표준화해야 할 것과 원핫 인코딩을 각각 이용해야 할 피처들이 많았는데, 이럴 때는 함수를 새로 정의해 전처리 함수를 만드는 것도 하나의 방법일 수 있겠다는 점을 배웠다.
