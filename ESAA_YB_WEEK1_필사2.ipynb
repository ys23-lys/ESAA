## DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

DataFrame: 칼럼명 有 (<-> 리스트, 넘파이 ndarray)

DataFrame으로 변환 시 칼럼명 지정 필요. : **pd.DataFrame(data,columns)**
- data: 리스트나 딕셔너리 또는 넘파이 ndarray를 입력.
- columns: 칼럼명 리스트를 입력.

```python
import numpy as np

col_name1=['col1'] # 칼럼명 리스트
list1=[1,2,3] # 리스트
array1=np.array(list1) # 넘파이 ndarray
print('array1 shape:',array1.shape)

# 리스트를 이용해 DataFrame 생성.
df_list1=pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame;\n',df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성.
df_array1=pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrmae:\n',df_array1)
```

```python
# 3개의 칼럼명이 필요함.
col_name2=['col1','col2','col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2=[[1,2,3],
       [11,12,23]]
array2=np.array(list2)
print('array2 shape:',array2.shape)
df_list2=pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n',df_list2)
df_array2=pd.DataFrame(array2,columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n',df_array2)
```

```python
# Key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑.
dict={'col1':[1,11],'col2':[2,22],'col3':[3,33]}
df_dict=pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n',df_dict)
```

### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

DataFrame -> ndarray

: array 만들기 **df_dict.values**

DataFrame -> 리스트

: array에 tolist() 호출 **df_dict.values.tolist()**

DataFrame -> 딕셔너리

: DataFrame에 to_dict() 호출 후 인자로 'list' 입력 **df_dict.to_dict('list')**

```python
# DataFrame을 ndarray로 변환.
array3=df_dict.values
print('df_dict.values 타입:',type(array3),'df_dict.values shape:',array3.shape)
print(array3)
```

```python
# DataFrame을 리스트로 변환.
list3=df_dict.values.tolist()
print('df_dict.values.tolist() 타입:',type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환.
dict3=df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:',type(dict3))
print(dict3)
```

## DataFrame 데이터 삭제

drop() 메서드 이용.

DataFrame.drop(**labels=None,axis=0**,index=None,columns=None,level=None,**inplace=False**,errors='raise')

**labels**: 여러 개의 칼럼을 삭제하고 싶을 경우 **리스트** 형태로 삭제하고자 하는 칼럼명을 입력.

**axis**: 특정 칼럼 또는 특정 행을 드롭.
- axis=1 입력 시 **칼럼 드롭** (기존 칼럼 값을 가공 > 새로운 칼럼 생성, 삭제)
- axis=0 입력 시 로우 드롭 (이상치 데이터를 삭제하는 경우)

**inplace**
- 디폴트값 False: 원본 DataFrame 유지, 드롭된 DataFrame을 새롭게 객체 변수로 받을 경우.
- **inplace=True** 원본 DataFrame에 드롭된 결과를 적용할 경우.

```python
# 'Age_0' 칼럼 삭제.
titanic_drop_df=titanic_df.drop('Age_0',axis=1)
titanic_drop_df.head(3)
```

```python
titanic_df.head(3) # inplace=False
```

```python
drop_result=titanic_df.drop(['Age_0','Age_by_10','Family_No'],axis=1,inplace=True)
print('inplace=True로 drop 후 반환된 값:',drop_result)
titanic_df.head(3)

# titanic_df=titanic_df.drop(['Age_0','Age_by_10','Family_No'],axis=1,inplace=True)는 titanic_df의 객체 변수를 None으로 만듦.
```

### DataFrame iloc[] 연산자
: row, col을 지정하여 데이터를 선택할 수 있는 인덱싱 방식.

**iloc[]**: 위치 기반 인덱싱 방식.
- 행과 열 위치를, 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값(또는 정수형의 슬라이싱, 팬시 리스트 값)으로 지정하는 방식.
- **iloc[행 위치 정숫값, 열 위치 정숫값]**
- **열 위치에 -1 입력**: 마지막 열 데이터 가져옴. (iloc[:,-1]: 맨 마지막 칼럼 값, iloc[:,:-1]: 맨 마지막 칼럼 값만 제외한 모든 값.)

loc[]: 명칭 기반 인덱싱 방식.
- 인덱스값으로 행 위치를, 칼럼의 명칭으로 열 위치를 지정하는 방식.

```python
data={'Name':['Chulmin','Eunkyung','Jinwoong','Soobeom'],
      'Year':[2011,2016,2015,2015],
      'Gender':['Male','Female','Male','Male']}
data_df=pd.DataFrame(data,index=['one','two','three','four'])
data_df
```

```python
data_df.iloc[0,0] # iloc[행 위치 정숫값, 열 위치 정숫값]
```

```python
# 아래 코드는 오류를 발생시킵니다.
data_df.iloc[0,'Name']
```

```python
# 아래 코드는 오류를 발생시킵니다.
data_df.iloc['one',0]
```

```python
data_df.iloc[1,0] # 2 row 1 col의 단일값 반환.
```

```python
data_df.iloc[2,1] # 3 row 2 col의 단일값 반환.
```

```python
data_df.iloc[0:2,[0,1]]
# 0:2 슬라이싱 범위의 1, 2 row와 1, 2 col의 DataFrame 반환.
```

```python
data_df.iloc[0:2,0:3]
# 0:2 슬라이싱 범위의 1, 2 row와 1, 2, 3 col의 DataFrame 반환.
```

```python
data_df.iloc[:] # 전체 DataFrame 반환.
```

```python
print("\n 맨 마지막 칼럼 데이터 [:,-1] \n",data_df.iloc[:,-1])
print("\n 맨 마지막 칼럼을 제외한 모든 데이터 [:,:-1] \n",data_df.iloc[:,:-1])
```

## 정렬, Aggregation 함수, GroupBy 적용

### DataFrame, Series의 정렬 - sort_values()

sort_values()

**by**: 특정 칼럼 입력 시 해당 칼럼으로 정렬 수행.

**ascending**=True(기본값): 오름차순 정렬.

## apply lambda 식으로 데이터 가공

lambda: 함수의 선언과 함수 내의 처리를 한 줄의 식으로 변환하는 식.

map(): lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우.

```python
def get_square(a):
  return a**2

print('3의 제곱은:',get_square(3))
```

```python
lambda_square=lambda x : x**2
print('3의 제곱은:',lambda_square(3))
```

```python
a=[1,2,3]
squares=map(lambda x : x**2,a)
list(squares)
```

```python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
```

if 절의 경우 if 식보다 반환값을 먼저 기술.

**lambda x: 'if일 때의 반환값' if 조건 else 'else일 때의 반환값'**

```python
# 나이가 15세 미만이면 Child, 아니면 Adult
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
# else if 사용법
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat=''
  if age<=5: cat='Baby'
  elif age<=12: cat='Child'
  elif age<=18: cat='Teenager'
  elif age<=25: cat='Student'
  elif age<=35: cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat='Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환.
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
```
