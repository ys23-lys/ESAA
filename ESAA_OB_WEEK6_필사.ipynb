## 4.2 딥러닝 구조

### 4.2.1 딥러닝 용어

층
- 입력층(input layer): 데이터를 받아들이는 층
- 은닉층(hidden layer): 모든 입력 노드부터 입력 값을 받아 가중합을 계산하고, 이 값을 활성화 함수에 적용하여 출력층에 전달하는 층
- 출력층(output layer): 신경망의 최종 결괏값이 포함된 층

가중치(weight): 노드와 노드 간 연결 강도. 입력 값이 연산 결과에 미치는 영향력을 조절하는 요소.

바이어스(bias): 가중합에 더해 주는 상수. 하나의 뉴런에서 활성화 함수를 거쳐 최종적으로 출력되는 값을 조절하는 역할을 함

가중합(weighted sum), 전달함수: 가중치와 신호의 곱을 합한 것

함수
- 활성화 함수(activation function): 신호를 입력받아 이를 적절히 처리하여 출력해 주는 함수
   - 시그모이드 함수: 선형 함수의 결과를 0~1 사이에서 비선형 형태로 변형.
   - 하이퍼볼릭 탄젠트 함수: 선형 함수의 결과를 -1~1 사이에서 비선형 형태로 변형
   - 렐루 함수: 입력이 음수일 때는 0 출력, 양수일 때는 x 출력
   - 리키 렐루 함수: 입력 값이 음수이면 0이 아닌 매우 작은 수 반환
   - 소프트맥스 함수: 입력 값을 0~1 사이에 출력되도록 정규화하여 출력 값들의 총합이 항상 1이 되도록 함
- 손실 함수(loss function): 가중치 학습을 위해 출력 함수의 결과와 실제 값 간의 오차를 측정하는 함수
   - 평균 제곱 오차
   - 크로스 엔트로피 오차: 분류 문제에서 원-핫 인코딩을 했을 때 사용할 수 있는 오차 계산법

### 4.2.2 딥러닝 학습

순전파(feedforward): 모든 뉴런이 이전 층의 뉴런에서 수신한 정보에 변환을 적용하여 다음 층의 뉴런으로 전송하는 방식

역전파(backpropagation): 손실 함수 비용이 0에 가깝도록 하기 위해 모델이 훈련을 반복하며 가중치를 조정하고 손실이 계산되면 그 정보가 역으로 전파됨(출력층 -> 은닉층 -> 입력층)

### 4.2.3 딥러닝의 문제점과 해결 방안

과적합 문제 발생: 드롭아웃으로 해결

기울기 소멸 문제 발생: 은닉층이 많은 신경망에서 주로 발생. 출력층에서 은닉층으로 전달되는 오차가 크게 줄어들어 학습이 되지 않는 현상.

성능이 나빠지는 문제 발생: 확률적 경사 하강법, 미니 배치 경사 하강법으로 해결
- 배치 경사 하강법(BGD): 전체 데이터셋에 대한 오류를 구한 후 기울기를 한 번만 계산하여 모델의 파라미터를 업데이트하는 방법
- 확률적 경사 하강법(SGD): 임의로 선택한 데이터에 대해 기울기를 계산하는 방법
- 미니 배치 경사 하강법: 전체 데이터셋을 미니 배치 여러 개로 나누고 미니 배치 한 개마다 기울기를 구한 후 그것의 평균 기울기를 이용하여 모델을 업데이트해서 학습하는 방법

옵티마이저: 확률적 경사 하강법의 파라미터 변경 폭이 불안정한 문제를 해결하기 위한 방법
- 속도를 조정하는 방법
   - 아다그라드(Adagrad): 변수의 업데이트 횟수에 따라 학습률을 조정하는 방법. 많이 변화하지 않는 변수들의 학습률은 크게, 많이 변화하는 변수들의 학습률은 작게 함.
   - 아다델타(Adadelta): 아다그라드에서 G값이 커짐에 따라 학습이 멈추는 문제를 해결하기 위해 등장.
   - 알엠에스프롭(RMSProp): 아다그라드의 G(i) 값이 무한히 커지는 것을 방지하고자 제안된 방법
- 운동량을 조정하는 방법
   - 모멘텀(Momentum): 가중치를 수정하기 전에 이전 수정 방향을 참고하여 같은 방향으로 일정한 비율만 수정하는 방법
   - 네스테로프 모멘텀(NAG): 모멘텀 값이 적용된 지점에서 기울기 값을 계산.
- 속도와 운동량에 대한 혼용 방법
   - 아담(Adam): 모멘텀과 알엠에스프롭의 장점을 결합한 경사 하강법
